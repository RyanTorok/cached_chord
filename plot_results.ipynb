```python
directory = '/Users/sayash/Downloads/experiments/nodes_sim_rtt60/'

df = pd.read_csv(directory+'stats.csv', header=None)
df
```

```python
df.columns=['latency','lookups','strategy','size', 'dist', 'nodes']
```

```python
df['latency'] = df['latency'].str[1:].astype(float)
df['nodes'] = df['nodes'].str[:-1].astype(int)
```

```python
df.head()
```

```python
df_strategy = df.groupby(['strategy','dist', 'size','nodes']).agg({'latency' : ['mean', 'std'], 'lookups' : ['mean', 'std']})
```

```python
df_strategy.reset_index()
```

```python
df_strategy.columns = df_strategy.columns.map('|'.join).str.strip('|')
```

```python
df_final=df_strategy.reset_index()
```
